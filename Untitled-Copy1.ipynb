```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,f1_score
```

```python
models=[0]*5
```

```python
models[0]=LogisticRegression()
models[1]=KNeighborsClassifier()
models[2]=DecisionTreeRegressor()
models[3]=DecisionTreeClassifier()
models[4]=SVC()
```

```python
models
```

```python
def check():
    for m in models:
        model=m
        model.fit(X_test,y_test)
        y_pred = model.predict(data.drop(columns=['target']))
        print('Model Name: ',m)
        print("Accuracy score is: ",accuracy_score(data['target'],y_pred)*100)
        print("F1 score is: ",f1_score(data['target'],y_pred)*100)
        print()
```

```python
check()
```

```python
from sklearn.model_selection import cross_val_score

for m in models:
    scores = cross_val_score(m, X_train, y_train, cv=10)*100
    print('Model Name: ',m)
    print('Cross-Validation Accuracy Scores', scores)
```
