# Seperating Numerical and Categorical Features

```python
num_cols=['lead_time','arrival_date_month','arrival_date_day_of_month','stays_in_weekend_nights','stays_in_week_nights','is_repeated_guest',
         'previous_cancellations','previous_bookings_not_canceled','days_in_waiting_list','adr']
cat_cols=['hotel','market_segment','distribution_channel','reserved_room_type','assigned_room_type','deposit_type','customer_type','reservation_status']
target_col = ['is_canceled']
```

<h4>Storing in new datafram</h4>

```python
new = df[num_cols + cat_cols + target_col]
```

```python
new.info()
```

<h4>Making a copy of new dataframe as data</h4>

```python
new.to_csv('data.csv')
```

```python
data=new.copy()
```

# Splitting the dataset in to train and test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(data.drop(columns=['is_canceled']),data['is_canceled'],test_size=0.25,random_state=20)
```

# Building the Model 
# Models used to predict :
<ol>
<li>LogisticRegression</li>
<li>KNeighborsClassifier</li>
<li>DecisionTreeClassifier</li>
<li>SVC</li>
</ol>

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,f1_score
```

Creating a list of models

```python
models=[0]*4
```

```python
models[0]=LogisticRegression()
models[1]=KNeighborsClassifier()
models[2]=DecisionTreeClassifier()
models[3]=SVC()
```

```python
models
```

<h3>Creating a function to check the accuracy score and f1 score for each model</h3>

```python
def check():
    for m in models:
        model=m
        model.fit(X_train,y_train)
        y_pred = model.predict(X_test)
        print('Model Name: ',m)
        print("Accuracy score is: ",accuracy_score(y_test,y_pred)*100)
        print("F1 score is: ",f1_score(y_test,y_pred)*100)
#         if m in [DecisionTreeRegressor(),DecisionTreeClassifier()]:
#             continue
#         print("Probability is : ",model.predict_proba(data.drop(columns=['target'])))
        print()
```

```python
check()
```

# Checking that our model is overfitted or not
# by fitting the test dataset and then predicting the train dataset

```python
def check():
    for m in models:
        model=m
        model.fit(X_test,y_test)
        y_pred = model.predict(X_train)
        print('Model Name: ',m)
        print("Accuracy score is: ",accuracy_score(y_train,y_pred)*100)
        print("F1 score is: ",f1_score(y_train,y_pred)*100)
#         if m in [DecisionTreeRegressor(),DecisionTreeClassifier()]:
#             continue
#         print("Probability is : ",model.predict_proba(data.drop(columns=['target'])))
        print()
```

```python
check()
```

<h3>Cross validating the models</h3>

```python
from sklearn.model_selection import cross_val_score

for m in models:
    scores = cross_val_score(m, X_train, y_train, cv=10)*100
    print('Model Name: ',m)
    print('Cross-Validation Accuracy Scores', scores)
    print()
```

<h3>Confusion Matrix for logistic regression</h3>

```python
model = LogisticRegression()
model.fit(X_test,y_test)
y_pred = model.predict(X_train)
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train,y_pred,normalize='true')*100
```
